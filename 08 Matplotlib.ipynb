```python
import matplotlib.pyplot as plt
```

```python
#plots things in the jupyter notebook
%matplotlib inline  
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#going to show a functional method and an object oriented method
```

```python
plt.plot(x,y, 'r-')
# plt.show()    use this for spyder. Very similar to print()
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
#multiplots 
```

```python
plt.subplot(1,2,1) #rows, columns, this plot index
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented method
```

```python
fig = plt.figure()   #creates a figure object

axis = fig.add_axes([0.1, 0.1,  0.8, 0.8]) #pass in a list. Left bottom width height

axis.plot(x,y)
axis.set_xlabel('X label')
axis.set_title('Title')
axis.set_ylabel('Y')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.15, 0.4, 0.4, 0.4]) #width and height are not percentages

axes1.plot(x,y)
axes1.set_title('big plot')

axes2.plot(y,x)
axes2.set_title('mini plot')
```

```python
#43 matplot lib part 2
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)  #the comma is touple unpacking
plt.tight_layout() #prevents overlap
#axes.plot(x,y)

#for current_ax in axes:  #axes is an object you can iterate through
#    current_ax.plot(x,y)
    
axes[0].plot(x,y)  #you can also access individual plots based on their position within the axes array
axes[0].set_title('first')
```

```python
#Figure size and DPI
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)  #pass in a touple of width and height in inches
#second arg is dpi. dots per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#legends  label which plot is which
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x,x**2, label='X squared')
ax.plot(x,x**3, label='X cubed')

ax.legend()  #if you just run this you get a warning saying no labelled objects found. 
#Must pair with the label = ''  during plotting
ax.legend(loc=1)
```

```python
#plot appearance 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='red', linewidth=3, alpha=0.5, linestyle = '-.') 
#can also use RGB hex code for color. Look online for custom hex codes
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='red', linewidth=3, alpha=0.5, marker='o', markersize=10) 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='red', linewidth=3, alpha=0.5, marker='o', markersize=10) 
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
