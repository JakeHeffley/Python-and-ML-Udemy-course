```python
#data 
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2  = pd.DataFrame({'Category':['A','B','C'], 'Values':[32,43,50]})
```

```python
df2.head()
```

```python
df.iplot() #interactive plotly plot
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.count().iplot(kind='bar') #can combine with aggregate functions
```

```python
df.iplot(kind='box') #box plot automatically aggregates data. 

#can click on the legend to toggle inclusion of boxes
```
