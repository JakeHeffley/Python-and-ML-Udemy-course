```python
# PART TWO - Data Preprocessing and Creating a Model
```

```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
#Base your number of units off the number of features
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, 
          validation_data=(X_test,y_test),
         batch_size=128, epochs=400) 
#after each data run the model on the validation data to track how things are 
#going. validation data will not affect the weights or biases updates.

#used batches because it is a larger dataset. 
```

```python
#PART THREE - Model Evaluation and Predictions
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot() #compare training loss to validation loss. Once there is an increase in validation loss 
#you have begun over fitting to the training data. 
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df['price'].describe()  #typical home price is 540K and we are off by <100K. Thats not good. 
```

```python
explained_variance_score(y_test, predictions) #tells you how much variance is actually explained by your model.
#in this case its around 80%
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test, predictions, alpha=0.3)
plt.plot(y_test, y_test, 'r')

#outliers seem to be causing a lot of error. Perhpas we should retrain while excluding top 1%
```

```python
#Retrain while dropping top 1%
X = non_top_1.drop('price', axis=1).values
y = non_top_1['price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, 
          validation_data=(X_test,y_test),
         batch_size=128, epochs=400) 
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
predictions = model.predict(X_test)
```
