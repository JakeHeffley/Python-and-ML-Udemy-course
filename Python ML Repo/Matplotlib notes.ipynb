```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
#Functional versus object oriented
```

```python
#FUNCTIONAL
```

```python
plt.plot(x,y, 'r--')
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT ORIENTED PLOTTING
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('XXXXXX')
axes.set_ylabel('YYYYY')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('smaller plot')
axes1.set_title('larger plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_title('First')
```

```python
#fig size, aspect ratio, plot size
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,3))

#ax = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='x sqared')
ax.plot(x,x**3, label = 'x cubed')

ax.legend(loc=0)
```

```python
#setting colors etc
```

```python
fig = plt.figure()

ax  = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='brown',linewidth=5, linestyle='--')
ax.plot(y,x,color='blue', linewidth=2, alpha=0.5, linestyle='-.')
```

```python
fig = plt.figure()

ax  = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='brown',lw=3, ls='--')
ax.plot(y,x,color='blue', lw=2, alpha=0.5, marker='o', markersize=10,
       markerfacecolor='yellow',markeredgewidth=3, markeredgecolor='green')
```

```python
#control over axis appearance and plot range
```

```python
fig = plt.figure()

ax  = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='brown',lw=3, ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
